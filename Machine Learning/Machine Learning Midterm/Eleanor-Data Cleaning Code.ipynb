```python
import pandas as pd
import numpy as np

# ---------------------------
# 0) Normalize key columns
# ---------------------------
cols = df.columns.tolist()
if "Country" not in cols and "Entity" in cols:
    df = df.rename(columns={"Entity": "Country"})
if "Code" in df.columns and "Country_Code" not in df.columns:
    df = df.rename(columns={"Code": "Country_Code"})

# ---------------------------
# 1) Create a safe rename for the outdoor air pollution column
#    (Use substring matching instead of giant exact strings)
# ---------------------------
def find_col(substring, cols):
    matches = [c for c in cols if substring.lower() in c.lower()]
    return matches[0] if matches else None

col_outdoor_long = find_col("Risk: Outdoor air pollution", df.columns)
col_total_air_long = find_col("Risk: Air pollution - Sex: Both", df.columns)  # sometimes present
col_household_long = find_col("Household air pollution", df.columns)          # may not be in this file

rename_map = {}
if col_outdoor_long:
    rename_map[col_outdoor_long] = "Deaths_due_to_Outdoor_Air_Pollution"
if col_total_air_long:
    rename_map[col_total_air_long] = "Deaths_due_to_Air_Pollution_Total"
if col_household_long:
    rename_map[col_household_long] = "Deaths_due_to_Household_Air_Pollution"

df = df.rename(columns=rename_map)

# ---------------------------
# 2) Sanity checks
# ---------------------------
required = ["Country", "Year"]
missing_req = [c for c in required if c not in df.columns]
if missing_req:
    raise KeyError(f"Missing required columns {missing_req}. Current columns: {df.columns.tolist()}")

target_col = "Deaths_due_to_Outdoor_Air_Pollution"
if target_col not in df.columns:
    raise KeyError(
        f"Could not find/rename the outdoor air pollution column. "
        f"Check the substring used in find_col() or print(df.columns.tolist())."
    )

# ---------------------------
# 3) Dedupe and coerce types
# ---------------------------
df = df.drop_duplicates(subset=["Country", "Year"]).copy()
df["Year"] = pd.to_numeric(df["Year"], errors="coerce")

# ensure numeric for the target before interpolation
df[target_col] = pd.to_numeric(df[target_col], errors="coerce")

# ---------------------------
# 4) Interpolate by country (sorted)
# ---------------------------
df = df.sort_values(["Country", "Year"])
df[target_col] = (
    df.groupby("Country", group_keys=False)[target_col]
      .apply(lambda s: s.interpolate(limit_direction="both"))
)

# Optional: also interpolate other renamed columns if they exist
for maybe_col in ["Deaths_due_to_Air_Pollution_Total", "Deaths_due_to_Household_Air_Pollution"]:
    if maybe_col in df.columns:
        df[maybe_col] = pd.to_numeric(df[maybe_col], errors="coerce")
        df[maybe_col] = (
            df.groupby("Country", group_keys=False)[maybe_col]
              .apply(lambda s: s.interpolate(limit_direction="both"))
        )

# ---------------------------
# 5) Drop any remaining rows missing the target
# ---------------------------
df = df.dropna(subset=[target_col]).reset_index(drop=True)

# ---------------------------
# 6) (Optional) Create per-100k rate if 'Population' exists
# ---------------------------
if "Population" in df.columns:
    df["Outdoor_Air_Pollution_per_100k"] = (df[target_col] / df["Population"]) * 100_000

# ---------------------------
# 7) Save only if file does not already exist
# ---------------------------
import os
output_path = r"C:\Users\elean\OneDrive - University of Virginia\Documents\DS 3001\Midterm\cleaned_air_pollution_deaths.csv"
if not os.path.exists(output_path):
    df.to_csv(output_path, index=False)
    print(f"✅ Cleaned dataset saved to: {output_path}")
else:
    print(f"⚠️ File already exists at: {output_path} (no overwrite)")

# ---------------------------
# 8) Quick integrity checks
# ---------------------------
print("Rows:", len(df))
print("Countries:", df["Country"].nunique())
print("Years:", int(df["Year"].min()), "→", int(df["Year"].max()))
print("Missing % in target:", round(df[target_col].isna().mean()*100, 3), "%")
```

```python
import pandas as pd
from scipy.stats import linregress

# Step 1: Create an empty list to hold results
summary_data = []

# Step 2: Loop through each country and calculate stats
for country, group in df.groupby("Country"):
    group = group.sort_values("Year")
    start_val = group.iloc[0]["Deaths_due_to_Outdoor_Air_Pollution"]
    end_val = group.iloc[-1]["Deaths_due_to_Outdoor_Air_Pollution"]
    avg_val = group["Deaths_due_to_Outdoor_Air_Pollution"].mean()
    
    # Calculate % change and slope of trend line
    pct_change = ((end_val - start_val) / start_val * 100) if start_val != 0 else None
    
    slope, intercept, r_value, p_value, std_err = linregress(
        group["Year"], group["Deaths_due_to_Outdoor_Air_Pollution"]
    )
    
    summary_data.append({
        "Country": country,
        "Start_Deaths_1990": round(start_val, 2),
        "End_Deaths_2019": round(end_val, 2),
        "Average_Deaths": round(avg_val, 2),
        "Absolute_Change": round(end_val - start_val, 2),
        "Percent_Change": round(pct_change, 2) if pct_change is not None else None,
        "Trend_Slope": round(slope, 2),  # annual change in deaths
        "R_squared": round(r_value ** 2, 3)  # goodness of fit
    })

# Step 3: Convert results to a DataFrame
summary_df = pd.DataFrame(summary_data)

# Step 4: Sort by Trend_Slope or Percent_Change (optional)
summary_df = summary_df.sort_values(by="Trend_Slope", ascending=False)

# Step 5: Display first few rows
print("✅ Summary of Air Pollution Deaths (1990–2019):")
display(summary_df.head(10))
```

```python
import os
import pandas as pd

# === 1) File paths (edit these if your files live elsewhere) ===
cleaned_path = r"C:\Users\elean\OneDrive - University of Virginia\Documents\DS 3001\Midterm\cleaned_air_pollution_deaths_final_named.csv"
rate_path    = r"C:\Users\elean\OneDrive - University of Virginia\Documents\DS 3001\Midterm\death-rate-from-air-pollution-per-100000.csv"

# Where to save the merged file
output_dir   = r"C:\Users\elean\OneDrive - University of Virginia\Documents\DS 3001\Midterm"
output_path  = os.path.join(output_dir, "merged_air_pollution_with_rates.csv")
os.makedirs(output_dir, exist_ok=True)

# === 2) Load ===
cleaned = pd.read_csv(cleaned_path)
rate    = pd.read_csv(rate_path)

# Strip stray whitespace from headers (defensive)
cleaned.columns = cleaned.columns.str.strip()
rate.columns    = rate.columns.str.strip()

# === 3) Normalize key columns: Country / Entity; Code / Country_Code ===
# (Your cleaned file typically uses Country/Country_Code; rate often uses Entity/Code)
def normalize_country_cols(df):
    cols = df.columns
    if "Country" not in cols and "Entity" in cols:
        df = df.rename(columns={"Entity": "Country"})
    if "Country_Code" not in cols and "Code" in cols:
        df = df.rename(columns={"Code": "Country_Code"})
    return df

cleaned = normalize_country_cols(cleaned)
rate    = normalize_country_cols(rate)

# Sanity check: we need Country, Country_Code, Year to merge cleanly
for c in ["Country", "Country_Code", "Year"]:
    if c not in cleaned.columns:
        raise KeyError(f"Missing '{c}' in cleaned file columns: {cleaned.columns.tolist()}")
    if c not in rate.columns:
        raise KeyError(f"Missing '{c}' in rate file columns: {rate.columns.tolist()}")

# === 4) Rename the rate column for clarity ===
rate_col_long = "Deaths - Cause: All causes - Risk: Air pollution - Sex: Both - Age: Age-standardized (Rate)"
if rate_col_long not in rate.columns:
    # try a fuzzy match just in case headers vary slightly
    candidates = [c for c in rate.columns if "Risk: Air pollution" in c and "Age-standardized" in c and "(Rate)" in c]
    if candidates:
        rate_col_long = candidates[0]
    else:
        raise KeyError("Could not find the air-pollution rate column in the rate file.")

rate = rate.rename(columns={rate_col_long: "Air_Pollution_Deaths_per_100k"})

# === 5) Merge on Country, Country_Code, Year ===
merged = cleaned.merge(
    rate[["Country", "Country_Code", "Year", "Air_Pollution_Deaths_per_100k"]],
    on=["Country", "Country_Code", "Year"],
    how="left"
)

# === 6) Keep only the columns we’ll analyze ===
drivers_cols = [
    "Deaths_due_to_Household_Air_Pollution",
    "Deaths_due_to_Smoking",
    "Deaths_due_to_High_Blood_Pressure",
    "Deaths_due_to_Low_Physical_Activity",
    "Deaths_due_to_Alcohol_Use",
]
base_cols = ["Country", "Country_Code", "Year", "Deaths_due_to_Air_Pollution_Total", "Air_Pollution_Deaths_per_100k"]
keep_cols = base_cols + [c for c in drivers_cols if c in merged.columns]

df = merged[keep_cols].dropna(subset=["Air_Pollution_Deaths_per_100k"])
df = df.sort_values(["Country", "Year"]).reset_index(drop=True)

print("Data shape after merge/filter:", df.shape)
print(df.head())

# === 7) Save merged ===
df.to_csv(output_path, index=False)
print(f"✅ Merged dataset saved to: {output_path}")
```

```python
# --- Trend metrics using ONLY material covered in class ---
# For each country: compute regression slope & p-value (rate ~ Year)
# and Pearson correlation (Year vs. rate), plus % change.

import numpy as np
import pandas as pd
from scipy.stats import linregress, pearsonr
from IPython.display import display

rows = []
for country, g in df.groupby("Country"):
    g = g.dropna(subset=["Air_Pollution_Deaths_per_100k"]).sort_values("Year")
    if len(g) < 3:
        continue  # need enough points to say anything meaningful

    years = g["Year"].values
    rate  = g["Air_Pollution_Deaths_per_100k"].values

    # 1) Linear trend via simple regression (what you learned in class)
    lr = linregress(years, rate)
    slope = lr.slope            # change in deaths per 100k per YEAR
    p_lin = lr.pvalue           # p-value for the slope (is the trend significant?)

    # 2) Pearson correlation between Year and the rate (learned in class)
    r, p_r = pearsonr(years, rate)

    # 3) % change from first to last year (intuitive magnitude)
    first = rate[0]
    last  = rate[-1]
    pct_change = np.nan
    if np.isfinite(first) and first != 0:
        pct_change = 100 * (last - first) / first

    rows.append({
        "Country": country,
        "n_years": int(g["Year"].nunique()),
        "year_min": int(g["Year"].min()),
        "year_max": int(g["Year"].max()),
        "slope_per_year": slope,         # + means increasing each year
        "linreg_pvalue": p_lin,          # < 0.05 → statistically significant trend
        "pearson_r": r,                  # strength & direction of linear relationship
        "pearson_pvalue": p_r,           # < 0.05 → correlation is significant
        "pct_change_first_to_last": pct_change
    })

trend_table = pd.DataFrame(rows).sort_values("slope_per_year", ascending=False).reset_index(drop=True)

# Show top increases and decreases to discuss in slides
print("Top increases:")
display(trend_table.head(10).round(4))

print("Top decreases:")
display(trend_table.tail(10).round(4))
```
