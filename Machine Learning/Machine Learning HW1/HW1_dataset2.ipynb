```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\elean\OneDrive - University of Virginia\Documents\DS 3001\HW1_dataset_2.csv")
df.head()
```

```python
plt.hist(df.iloc[:,0], bins=20, edgecolor="black")
plt.title("Distribution of Random Numbers")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
### Distribution Modeling

#The histogram for this dataset shows a **bell-shaped, symmetric distribution** centered around 62â€“64.  
# - The frequencies decrease smoothly as values move away from the mean.  
# - The shape is consistent with the **Normal distribution**, which is characterized by symmetry and concentration around the mean.  

# Therefore, I believe the **Normal distribution** is an effective model for this dataset.
```

```python
import sys
!{sys.executable} -m pip install scipy

import numpy as np
from scipy.stats import norm

data = df.iloc[:,0]  # your random numbers column

# Estimate parameters
mu, sigma = np.mean(data), np.std(data, ddof=1)
print("Estimated mean (mu):", mu)
print("Estimated standard deviation (sigma):", sigma)
```

```python
# Histogram of the data (normalized to density)
plt.hist(data, bins=20, density=True, alpha=0.6, color="g", edgecolor="black")

# Normal distribution PDF with estimated mu and sigma
x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'k', linewidth=2, label="Normal fit")

plt.title(f"Normal Distribution Fit: mu={mu:.2f}, sigma={sigma:.2f}")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
