```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\elean\OneDrive - University of Virginia\Documents\DS 3001\HW1_dataset_1.csv")
df.head()
# head() shows the first 5 rows of the dataframe
```

```python
plt.hist(df.iloc[:,0], bins=20, edgecolor="black")
#hist() creates a histogram of the first column in the dataframe with 20 bins and black edges
plt.title("Distribution of Random Numbers")
#title() adds a title to the histogram
plt.xlabel("Value")
#xlabel() adds a label to the x-axis
plt.ylabel("Frequency")
#ylabel() adds a label to the y-axis
plt.show()
```

```python
### Distribution Modeling

#Based on the histogram, the feature appears to follow a **Normal distribution**.  
#- The data is centered around values between 5 and 7.  
#- The distribution is roughly **symmetric** with a bell-shaped curve.  
#- Frequencies decrease smoothly as we move away from the center in both directions.  

# These characteristics are consistent with the Normal distribution. Therefore, I believe that a **Normal distribution** models this feature effectively.
```

```python
import sys
!{sys.executable} -m pip install scipy

import numpy as np
from scipy.stats import norm

data = df.iloc[:,0]  # your random numbers column

# Estimate parameters
mu, sigma = np.mean(data), np.std(data, ddof=1)
print("Estimated mean (mu):", mu)
print("Estimated standard deviation (sigma):", sigma)
```

```python
# Histogram of the data (normalized to density)
plt.hist(data, bins=20, density=True, alpha=0.6, color="g", edgecolor="black")

# Normal distribution PDF with estimated mu and sigma
x = np.linspace(min(data), max(data), 100)
pdf = norm.pdf(x, mu, sigma)
plt.plot(x, pdf, 'k', linewidth=2, label="Normal fit")

plt.title(f"Normal Distribution Fit: mu={mu:.2f}, sigma={sigma:.2f}")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.show()
```
