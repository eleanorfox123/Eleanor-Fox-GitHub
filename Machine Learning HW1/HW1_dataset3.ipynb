```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Load the dataset
df3 = pd.read_csv(r"C:\Users\elean\OneDrive - University of Virginia\Documents\DS 3001\HW1_dataset_3.csv")

# 2. Check the first few rows to confirm columns
print(df3.head())

# 3. Create a scatter plot
plt.figure(figsize=(8,6))
plt.scatter(df3["X"], df3["Y"], alpha=0.7, edgecolor="k")

# 4. Add labels and title
plt.title("Scatter Plot of HW1_dataset3")
plt.xlabel("X")
plt.ylabel("Y")

plt.show()
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

X_vals = df3["X"].values.reshape(-1,1)
y_vals = df3["Y"].values

model = LinearRegression()
model.fit(X_vals, y_vals)
y_pred = model.predict(X_vals)

plt.figure(figsize=(8,6))
plt.scatter(df3["X"], df3["Y"], s=10, alpha=0.3, label="Data")
plt.plot(df3["X"], y_pred, color="red", linewidth=2, label="Regression Line")
plt.grid(True, linestyle="--", alpha=0.6)
plt.title("Scatter Plot with Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
### Scatter Plot Visualization

#The scatter plot shows the relationship between X and Y.  
#- Transparency and smaller point size were used to reduce overplotting and make dense areas more visible.  
#- Grid lines make it easier to interpret the spread of values.  
#- The dataset shows a clear positive relationship between X and Y, which can be highlighted further by fitting a regression line. 
## This ensures the visualization is both **informative** (showing the trend and spread of data) and **clear** (easy to interpret), meeting the requirements for clarity and effectiveness.
```
