```python
# 100% stacked bar chart: PRE vs POST (Matplotlib)
import matplotlib.pyplot as plt
import numpy as np

# --- Your data from the pies ---
categories = ["Bored", "Not great", "OK", "Kind of interested", "Excited"]
pre  = [11, 5, 40, 25, 19]   # PRE percentages
post = [12, 6, 14, 30, 38]   # POST percentages

# --- Build the chart ---
fig, ax = plt.subplots(figsize=(8, 5))

bottom_pre = 0
bottom_post = 0
legend_handles = []

for i, cat in enumerate(categories):
    b1 = ax.bar("Pre",  pre[i],  bottom=bottom_pre)
    b2 = ax.bar("Post", post[i], bottom=bottom_post)
    bottom_pre  += pre[i]
    bottom_post += post[i]
    # capture one patch per category for legend
    if i == 0:
        legend_handles.append(b1.patches[0])
    else:
        legend_handles.append(ax.containers[i].patches[0])

# Axes & labels
ax.set_ylim(0, 100)
ax.set_ylabel("Percent of responses (%)")
ax.set_title("Shift in attitudes about doing science after pilot program")

# Legend (uses the default Matplotlib colors)
ax.legend(legend_handles, categories, bbox_to_anchor=(1.03, 1), loc="upper left", title="Response")

# Optional: annotate segment percentages
# (labels centered within each stacked segment)
for container in ax.containers[:len(categories)]:  # "Pre" stacks
    for rect in container.patches:
        h = rect.get_height()
        if h >= 5:  # skip tiny labels
            ax.text(rect.get_x() + rect.get_width()/2, rect.get_y() + h/2,
                    f"{h:.0f}%", ha="center", va="center", fontsize=8)

for container in ax.containers[len(categories):]:  # "Post" stacks
    for rect in container.patches:
        h = rect.get_height()
        if h >= 5:
            ax.text(rect.get_x() + rect.get_width()/2, rect.get_y() + h/2,
                    f"{h:.0f}%", ha="center", va="center", fontsize=8)

plt.tight_layout()
plt.show()
```
