### Interactive Bar Graph using Plotly 

**A static graph, for comparison**

````python
# Create a new figure with a specific size (10 inches wide by 6 inches tall).
plt.figure(figsize=(10, 6))  # `figsize` specifies the dimensions of the figure in inches.

# Create a bar chart with sorted data from 'country_data_sorted'.
# 'country' column is used for x-axis labels, and 'population_millions' column is used for bar heights.
bars = plt.bar(
    country_data_sorted['country'], 
    country_data_sorted['population_millions'], 
    color='#fe5a03',  # Hexadecimal color for bars
    zorder=2  # Ensures the bars appear above gridlines or other background elements.
)

# Add a subtitle to describe what the chart shows.
plt.suptitle(
    'The total population in Brazil exceeds that of other countries', 
    x=0.32,  # Horizontal alignment of the title
    weight='bold'  # Bold text for emphasis
)

# Use the title as the y-axis label to indicate the unit of measurement.
plt.title(
    '(Millions of people)', 
    x=0.035,  # Horizontal alignment of the subtitle
    size=10  # Font size
)

# Rotate the x-axis labels slightly for better readability.
plt.xticks(rotation=0)  # `rotation=0` means no rotation. 

# Remove the gridlines to keep the chart clean.
plt.grid(False)  # Disables both x and y gridlines.

# Customize the y-axis tick marks to appear at intervals of 50 units.
plt.gca().yaxis.set_major_locator(plt.MultipleLocator(50))  
# `plt.gca()` gets the current axes object, allowing us to modify axis properties.
# `MultipleLocator(50)` ensures tick marks appear every 50 units on the y-axis.

# Remove the top and right borders (spines) of the plot for a cleaner appearance.
plt.gca().spines['top'].set_visible(False)  # Hides the top border.
plt.gca().spines['right'].set_visible(False)  # Hides the right border.

# Optionally remove the left border (spine) for an even cleaner look.
plt.gca().spines['left'].set_visible(False)  # Hides the left border.

# Remove the tick marks (lines) from the x-axis and y-axis to declutter the chart.
plt.gca().tick_params(axis='x', which='both', length=0)  # Removes x-axis tick lines.
plt.gca().tick_params(axis='y', which='both', length=0)  # Removes y-axis tick lines.

# Add labels over the bars.
for bar in bars:  # Loop through each bar object in the bar chart.
    yval = bar.get_height()  # Get the height of the current bar (value of the bar).
    plt.text(
        bar.get_x() + bar.get_width()/2,  # X-coordinate for the text (center of the bar).
        yval + 5,  # Y-coordinate slightly above the bar.
        round(yval),  # The value to display, rounded to the nearest integer.
        ha='center',  # Horizontal alignment: center the text.
        va='bottom'  # Vertical alignment: place text just above the bar.
    )

# Add a text box at the bottom left of the figure to credit the data source.
plt.figtext(
    0.1,  # X-coordinate for text placement in figure-relative coordinates.
    0.0,  # Y-coordinate for text placement in figure-relative coordinates.
    'Source: The World Bank',  # The text content.
    ha='left',  # Horizontal alignment: left-justify the text.
    fontsize=10  # Font size of the text.
)

# Finally, display the plot on the screen.
plt.show()  # Renders the plot in the output window or notebook.
````

````python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go

df = df.copy()
# Ensure Year sorts correctly (handles strings)
df['Year'] = pd.to_datetime(df['Year'].astype(str), format='%Y').dt.year

if 'Country' in df.columns:
    # Base figure: all countries, de-emphasized
    fig = px.line(
        df.sort_values(['Country','Year']),
        x='Year', y='Health_Care_Spending',
        color='Country', line_group='Country',
        hover_name='Country',
        template='plotly_white',
        labels={'Health_Care_Spending':'Health care spending', 'Year':'Year'}
    )
    for tr in fig.data:
        tr.update(line=dict(width=1), opacity=0.25)  # light, thin lines

    # Highlight a focal country (change name if you want a different focus)
    focus = 'United States'
    if (df['Country'] == focus).any():
        us = df[df['Country'] == focus].sort_values('Year')
        fig.add_trace(go.Scatter(
            x=us['Year'], y=us['Health_Care_Spending'],
            mode='lines+markers',
            name=focus,
            line=dict(width=4, color='#ffb503'),
            marker=dict(size=6),
            hovertemplate=f'{focus}<br>Year=%{{x}}<br>Spending=%{{y:,.1f}}<extra></extra>'
        ))
        # Hide the light duplicate of the focus line from px
        fig.for_each_trace(lambda t: t.update(visible='legendonly')
                           if t.name == focus and t.line.width == 1 else None)
else:
    # Fallback if you truly have just one series
    fig = px.line(df.sort_values('Year'), x='Year', y='Health_Care_Spending',
                  template='plotly_white', labels={'Health_Care_Spending':'Health care spending'})

# Nice touches
fig.update_layout(
    height=520, width=980,
    legend_title_text='Country',
    hovermode='x unified',
    margin=dict(l=40, r=20, t=40, b=40)
)
fig.update_xaxes(rangeslider_visible=True, showspikes=True)
fig.update_yaxes(tickformat=",.1f", zeroline=False)   # use "%, .1f%%" if it's a percent

fig.show()
````

#### General overview

In Plotly, `fig` is short for **Figure**  it's the **core object** that holds your entire chart, including:    
- Data  
- Layout (axes, labels, titles, etc.)   
- Traces (visual elements like lines, bars, points) 
- Interactivity 

**What does fig do?**
- Holds the data and layout for the chart 
Creates a chart (e.g., scatter, bar, etc.)  

Can be customized after creation using methods like:    

* `fig.update_layout()` 

* `fig.update_traces()` 

* `fig.add_trace()` (less common in Plotly Express, more in Graph Objects)  

Displays when you call `fig.show()` in a notebook or script 

**Customizing the Figure (fig)**  
  
`fig.update_layout()`   
Used to modify overall layout, such as: 

* Chart titles  

* Axis labels and ranges    

* Legend position   

* Templates and themes  

`fig.update_traces()`
Used to customize the traces (visual marks), such as:

* Marker size and color 

* Line style    

* Hover info
