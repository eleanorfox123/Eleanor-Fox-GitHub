 # Why is this an example of correlation, not causation?

This is an example of correlation because the increase in babys named eleanor throughout time is related to one another, but they do not cause the other to happen. There is likely a confounding variable in this graph like an increase in the amount of births over time which increases the amount of baby's. 


# Can you think of a confounding variable that might explain this correlation?

The confounding variable that I think would explain this correlation would be birth rate. As more and more people have babies, more babies need names and that is how the name Eleanor becomes more popular. There could also be cultural or trends that influence people to name their child Eleanor. 


# How does the axis scaling affect the visual impression?

The x-axis is distributed evenly throughout the graph. The y-axis starts close to the lowest value and increases by a great number (1.8k). If the y-axis is stretched out, the same increase looks much less dramatic. Axis scaling can either emphasize or downplay the trend. 


# Would a different axis scale change how strong the correlation looks?

Yes — while the mathematical correlation coefficient would stay the same, the visual impression of the relationship could look stronger or weaker depending on the scale. A tighter y-axis exaggerates the growth; a broader y-axis flattens it, making the correlation look weaker.
