```python
import pandas as pd
import matplotlib.pyplot as plt

# Your dataset
data = {
    'Country': ['U.S.', 'Germany', 'Japan', 'France', 'U.K.', 'Spain', 'Argentina', 'Brazil', 'Peru', 'Lebanon',
                'Indonesia', 'Ukraine', 'Morocco', 'Bangladesh', 'Nigeria', 'India', 'Kenya', 'Cameroon', 'Zimbabwe',
                'Ethiopia', 'Mali'],
    'Education': [20, 48, 10, 28, 18, 27, 51, 55, 60, 35, 29, 30, 33, 40, 28, 55, 26, 27, 29, 47, 50],
    'Hard work': [30, 18, 52, 28, 30, 40, 24, 26, 24, 20, 55, 26, 29, 25, 27, 25, 48, 28, 58, 42, 25],
    'Family wealth or connections': [30, 10, 20, 18, 27, 10, 12, 8, 5, 22, 17, 3, 14, 3, 22, 5, 20, 28, 10, 4, 15],
    'All, none or other': [10, 20, 18, 23, 18, 18, 5, 4, 2, 20, 5, 10, 15, 27, 15, 12, 5, 8, 1, 8, 10],
    'gni_per_capita': [65000, 48000, 42000, 41000, 40000, 32000, 20000, 15000, 13000, 12000,
                       11000, 9000, 8000, 7000, 6000, 5000, 4000, 3000, 2000, 1000, 500]
}
df = pd.DataFrame(data)

# Sort countries by GNI (richest to poorest)
df = df.sort_values("gni_per_capita", ascending=False)

# Set countries as index so they appear on y-axis
df = df.set_index("Country")

# Choose factor columns
factors = ['Education', 'Hard work', 'Family wealth or connections', 'All, none or other']
colors = ['#54728C', '#62A39F', '#915F84', '#C09550']

# Make subplots (no gaps between them)
fig, axes = plt.subplots(ncols=4, sharey=True, figsize=(12, 7), gridspec_kw={'wspace': 0})

for ax, factor, color in zip(axes, factors, colors):
    ax.barh(df.index, df[factor], color=color)
    
    # Bolded subplot titles
    ax.set_title(factor, fontsize=11, loc='left', weight='bold')
    
    # Different x-axis ranges depending on the factor
    if factor in ["Family wealth or connections", "All, none or other"]:
        ax.set_xlim(0, 50)   # only need up to 50%
        ax.set_xticks([0, 25, 50])
        ax.set_xticklabels(["0%", "25%", "0%"])
    else:
        ax.set_xlim(0, 100)  # keep full width for Education and Hard work
        ax.set_xticks([0, 25, 50, 75, 100])
        ax.set_xticklabels(["0%", "25%", "50%", "75%", "0%"])

    # Remove default x-axis grid
    ax.grid(False)

    # Draw vertical white lines ON TOP of bars
    for xpos in ([25, 50] if factor in ["Family wealth or connections", "All, none or other"] else [25, 50, 75]):
        ax.axvline(x=xpos, color="white", linewidth=1.2, zorder=5)

    # Horizontal gray lines for each country
    ax.yaxis.grid(True, color="lightgray", linewidth=0.8)
    ax.set_axisbelow(True)

    # Clean up spines
    for spine in ax.spines.values():
        spine.set_visible(False)


# Keep country names visible on the y-axis
axes[0].set_ylabel("")
axes[0].set_xlabel("Percentage")
for ax in axes[1:]:
    ax.set_ylabel("")
    ax.set_xlabel("Percentage")

# Main title (left aligned) + subtitle
fig.text(0.0, 0.98, 
         "What Young People Think is the Most Important Factor in Determining Success", 
         fontsize=16, ha="left", va="top")
fig.text(0.0, 0.94, 
         "How 15- to 24-year-olds in each country responded", 
         fontsize=11, ha="left", va="top")

# Sub text (multi-line paragraph at bottom)
fig.text(0.0, -0.02,
        "Note: Countries are sorted from richest to poorest according to a measure of gross national income per capita used by the\n" 
        "World Bank. Source: UNICEF-Gallup survey of more than 21,000 people conducted by telephone in 21 countries between February and June 2021.",
        fontsize=8, ha="left", va="top")

plt.tight_layout(rect=[0, 0, 1, 0.92])
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Your dataset
data = {
    "Term Start": [1917, 1919, 1921, 1923, 1925, 1927, 1929, 1931, 1933, 1935, 1937, 1939, 1941, 1943, 1945, 1947, 1949, 1951, 1953, 1955, 1957, 1959, 1961, 1963, 1965, 1967, 1969, 1971, 1973, 1975, 1977, 1979, 1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999, 2001, 2003, 2005, 2007, 2009, 2011, 2013, 2015, 2017],
    "Term End": [1919, 1921, 1923, 1925, 1927, 1929, 1931, 1933, 1935, 1937, 1939, 1941, 1943, 1945, 1947, 1949, 1951, 1953, 1955, 1957, 1959, 1961, 1963, 1965, 1967, 1969, 1971, 1973, 1975, 1977, 1979, 1981, 1983, 1985, 1987, 1989, 1991, 1993, 1995, 1997, 1999, 2001, 2003, 2005, 2007, 2009, 2011, 2013, 2015, 2017, 2019],
    "Women Representatives": [1, 1, 4, 2, 3, 5, 9, 8, 8, 6, 6, 9, 9, 9, 10, 10, 9, 10, 11, 17, 16, 16, 17, 12, 11, 11, 10, 13, 16, 19, 18, 16, 22, 22, 23, 24, 29, 32, 48, 50, 55, 58, 60, 62, 66, 74, 77, 73, 79, 84, 89],
    "Total Representatives": [435] * 51
}
df = pd.DataFrame(data)

# Calculate percentage of women
df["Percent Women"] = df["Women Representatives"] / df["Total Representatives"] * 100

# X labels: "start-end"
df["Term"] = df["Term Start"].astype(str) + "-" + df["Term End"].astype(str)

# Plot
fig, ax = plt.subplots(figsize=(10,6))

# Background gray bars (up to 25%)
ax.bar(df["Term"], [25]*len(df), color="lightgray")

# Orange bars for % women
ax.bar(df["Term"], df["Percent Women"], color="orange")

# Title
ax.set_title("The 116th Congress represents the biggest jump\nin women members since the 1990s", 
             fontsize=14, loc="left", weight="bold")

# Y-axis settings
ax.set_ylim(0, 50)
ax.set_yticks([0, 25, 50])
ax.set_yticklabels(["0%", "25%", "50%"])

# X-axis settings
ax.set_xticks(range(len(df)))
ax.set_xticklabels(df["Term"], rotation=60, ha="right", fontsize=8)

# Remove chart spines
for spine in ["top","right","left","bottom"]:
    ax.spines[spine].set_visible(False)

# Source note
fig.text(0.0, -0.05, "Source: Drew Desilver (2018)", fontsize=8, ha="left")

plt.tight_layout(rect=[0,0,1,0.95])
plt.show()
```

In this markdown cell, explain how the scaffolding of the graph helps to tell a story (including subtext). 

How the scaffolding of the graph tells a story

1. Gray background bars capped at 25%
- The background scaffolding sets a reference ceiling that makes the orange bars look small and struggling for most of the timeline.
- This visual contrast emphasizes how few women were in Congress for much of U.S. history.
- By stopping the gray scaffolding at 25%, the graph visually reinforces the idea that even today, representation remains below parity (50%).

2. Orange bars layered in front
- The bold orange color stands out against the muted gray scaffolding.
- This layering creates a story of change over time, with each orange bar climbing higher and encroaching on the gray background.

3. Left-aligned, bold title
- The title is phrased as a claim: â€œThe 116th Congress represents the biggest jump in women members since the 1990s.â€
- This scaffolds the viewerâ€™s interpretation, pointing attention specifically to the recent spike, rather than the entire timeline.

4. Y-axis labeled at 0%, 25%, and 50%
- The limited tick marks help readers quickly situate progress relative to parity.
- Having 50% visible but unattainable in the data adds subtext: even the historic 116th Congress is still far from equal representation.

5. X-axis with detailed term labels
- The rotated two-year term labels provide historical scaffolding, grounding the story in nearly a century of U.S. politics.
- The density of labels also signals continuity and slow change â€” until the sharp rise in recent decades.

6. Source note at the bottom
- The small citation to Drew Desilver (2018) signals credibility and transparency.
- This scaffolding reminds the reader that the graph is part of a larger narrative built on evidence.

Subtext

The scaffolding of the chart subtly conveys two messages:
- Historical inertia: The long stretch of small orange bars against a flat gray ceiling underscores how little changed for most of the 20th century.
- Recent acceleration, but unfinished story: The steep rise in the last few decades visually dramatizes progress, but the 50% tick mark looms above, reminding viewers that equality has not yet been achieved.
